# PROBLEMA 8.1

## Solución

Funcional de Costo:
min = 2500x1 + 1500x2 + 900x3

Restricciones:

80x1 + 60x2 + 40x3 >= 5000

70x1 + 50x2 + 30x3 >= 4000

65x1 + 40x2 + 20x3 >= 3000

```python
#Definir Matrices
from scipy.optimize import linprog
import numpy as np
c = np.array([2500, 1500, 900])
b = np.array([-5000,-4000,-3000])
A = np.array([[-80,-60,-40],
             [-70,-50,-30],
             [-65,-40,-20]])
```

```python
#Solucionar usando linprog

solucion = linprog(c, A_ub=A, b_ub=b)
solucion
```

# Resultado y Conclusión:

Gracias a la solución utilizando programación lineal, sabemos que la manera más óptima para minimizar los costos de los viajes es efectuar 50 viajes a la mina norte y a la mina central y ninguno a la mina sur. 

# Problema 8.2

## Solución

Funcional de Costo:
max = 1100xA + 1300xB + 1200xC + 980 xD + 1010xE

Restricciones:

3xA + 2xB + 5xC + 2xD + 1xE <= 17 

1xA + 3xB + 2xC + 4xD + 0xE <= 11

3xA + 2xB + 2xC + 1xD + 2xE <= 15

1xA + 2xB + 3xC + 1xD + 3xE <= 12

1xA + 0xB + 2xC + 2xD + 3xE <= 14

```python
#Definir Matrices

c = np.array([-1100,-1300,-1200,-980,-1010])
b = np.array([17,11,15,12,14,])
A = np.array([[3,2,5,2,1],
              [1,3,2,4,0],
              [3,2,2,1,2],
              [1,2,3,1,2],
              [1,3,1,2,3]])
```

```python
#Solucionar usando linprog

solucion = linprog(c, A_ub=A, b_ub=b)
solucion
```

# Resultado y Conclusión

Usando los resultados obtenidos con la programación lineal, podemos saber que la manera en que maximizamos la utilidad a partir de la venta de sillas es produciendo 2 Sillas A, 1 Silla B, 1 Silla C, 1 Silla D y 2 Sillas E.
